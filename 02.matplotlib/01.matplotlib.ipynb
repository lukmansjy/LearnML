```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100) # membuat element 0 - 10, dibagi sama rata 100
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x, y1)
ax.plot(x, y2)
```

## shortcut

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y1, "r-.", linewidth=0.5)
plt.plot(x, y2, "b-v", linewidth=0.5, markersize=3)
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title("Sin dan Cos")
plt.xlabel("x")
plt.ylabel("y")
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(x, y1, "r-.", linewidth=0.5)
ax.plot(x, y2, "b-v", linewidth=0.5, markersize=3)
ax.set_xlim(0, 10)
ax.set_ylim(-2, 2)
ax.set_title("Sin dan Cos", fontsize=14)
ax.set_xlabel("x", fontsize=10)
ax.set_ylabel("y", fontsize=10)
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(x, y1, "r-.", linewidth=0.5)
ax.plot(x, y2, "b-v", linewidth=0.5, markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 2), title="Sin dan Cos", xlabel="x", ylabel="y")
ax.axis('equal')
```

## Legend

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(x, y1, "r-.", linewidth=0.5, label="sin")
ax.plot(x, y2, "b-v", linewidth=0.5, markersize=3, label="cos")
ax.set(xlim=(0, 10), ylim=(-2, 2), title="Sin dan Cos", xlabel="x", ylabel="y")
ax.axis('equal')
ax.legend(loc="lower center", ncol=2, shadow=True, borderpad=1)
```

## Matplotlib Style

```python
fig, ax = plt.subplots(2, 3, figsize=(15,10))

ax[0, 0].hist(numbers, bins=80, density=1)

ax[0, 1].axis('off')

ax[0, 2].hist(numbers, bins=80, density=1)

ax[1, 0].axis('off')

ax[1, 1].scatter(y, x, c=rad, s=rad*50, cmap="seismic")
ax[1, 1].axis("equal")

ax[1, 2].hist(numbers, bins=80, density=1);
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15,10), sharex=True, sharey=True)

ax[0, 0].hist(numbers, bins=80, density=1)
ax[0, 2].hist(numbers, bins=80, density=1)
ax[1, 1].scatter(y, x, c=rad, s=rad*50, cmap="seismic")
ax[1, 2].hist(numbers, bins=80, density=1);
```

## 3D plot 

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
z = np.sqrt(X**2 + Y**2 + 10)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z, antialiased=False, cmap="winter")
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=10, azim=45)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, z, cmap="winter")
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-20, azim=45)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, z, 50, cmap="winter")
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-20, azim=45)
```
